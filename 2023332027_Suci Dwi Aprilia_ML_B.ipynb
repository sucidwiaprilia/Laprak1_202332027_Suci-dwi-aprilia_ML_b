## Scatter Plot

```python
# Umur vs. Probabilitas Botak
fig, ax = plt.subplots()
ax.scatter(df["umur"], df["botak_prob"], color='pink', alpha=0.5, label='Umur vs Probabilitas Botak')
ax.set_xlabel("Umur (tahun)")
ax.set_ylabel("Probabilitas Botak")
ax.set_title("Scatter Plot: Umur vs Probabilitas Botak")
ax.legend()
plt.show()
```

## Pie Chart

```python
# Distribusi Jenis Kelamin
fig, ax = plt.subplots()
gender_counts = df["jenis_kelamin"].value_counts()
ax.pie(gender_counts, labels=gender_counts.index, autopct='%1.2f%%', colors=["lightblue", "lightcoral"], startangle=90)
ax.set_title("Distribusi Jenis Kelamin")
plt.show()
```

## Box Pot

```python
# Pengaruh Merokok terhadap Kebotakan
fig, ax = plt.subplots(figsize=(8, 6))
filtered_df = df[df["is_merokok"].isin([0, 1])]
box = ax.boxplot([
    filtered_df[filtered_df["is_merokok"] == 0]["botak_prob"],  # Tidak Merokok
    filtered_df[filtered_df["is_merokok"] == 1]["botak_prob"]   # Merokok
], vert=False, showmeans=True, meanline=True, patch_artist=True,
   medianprops={'linewidth': 2, 'color': 'blue'},
   meanprops={'linewidth': 2, 'color': 'red'})
ax.set_yticklabels(['Tidak Merokok', 'Merokok'])
ax.set_ylabel("Kategori Merokok")
ax.set_title("Box Plot: Pengaruh Merokok terhadap Probabilitas Botak")
colors = ['lightblue', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.show()
```
